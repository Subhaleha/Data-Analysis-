```python
def area(radius):
    return 3.14 * radius * radius
r = float(input("Enter radius: "))
print("Area =", area(r))
```

```python
def largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))
print("Largest =", largest(x, y, z))
```

```python
def string_length(text):
    count = 0
    for i in text:
        count = count + 1
    return count
word = input("Enter a string: ")
print("Length =", string_length(word))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
num = int(input("Enter a number: "))
if is_prime(num):
    print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
n = int(input("Enter n: "))
print("Fibonacci =", fibonacci(n))
```

```python
def reverse_string(text):
    if text == "":
        return ""
    return reverse_string(text[1:]) + text[0]
word = input("Enter a string: ")
print("Reversed String:", reverse_string(word))
```
